```python
files = [file for file in os.listdir('./Sales_Data')]
all_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_data = pd.concat([all_data, df])
    
all_data.to_csv('all_months_data.csv',index=False)
```

```python
all_data = pd.read_csv('all_months_data.csv')
all_data.head(10)
```

```python
all_data = all_data.dropna(how='all')
all_data
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
all_data
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float')
all_data["Sales"] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
res1=all_data.groupby('Month').sum()
res1['Sales']
```

```python
months = range(1,13)
plt.bar(months,res1['Sales'])
plt.xticks(months)
plt.title("sales by month")
plt.xlabel("months")
plt.ylabel("Sales in $")
plt.show()
```

```python
def get_city(adress):
    return adress.split(',')[1]
    
def get_state(adress):
    return adress.split(',')[2].split(' ')[1]

all_data["City"] = all_data["Purchase Address"].apply(lambda x : f"{get_city(x)} ({get_state(x)})")
all_data
```

```python
res2 = all_data.groupby('City').sum()
res2
```

```python
cities = [city for city ,df in all_data.groupby('City')]


plt.bar(cities,res2['Sales'])
plt.xticks(cities, rotation="vertical", size=8)
plt.title("sales by city")
plt.xlabel("cities")
plt.ylabel("Sales in $")
plt.show()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('number of orders')
plt.grid()
plt.show()
```

```python
minutes = [minute for minute, df in all_data.groupby('Minute')]

plt.plot(minutes, all_data.groupby('Minute').count())
plt.xticks(minutes, size=5)
plt.xlabel('minutes')
plt.ylabel('Sales in $')
plt.grid()
plt.show()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key,value)
```
